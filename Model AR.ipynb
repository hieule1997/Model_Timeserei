```python
# AR example
from statsmodels.tsa.ar_model import AR
from random import random
# contrived dataset
data = [x +random() for x in range(1, 100)]
# fit model
# data.plot()
print(data)
import matplotlib.pyplot as plt
plt.plot(data)
plt.ylabel('some numbers')
plt.show()

model = AR(data)
model_fit = model.fit()
# make prediction
yhat = model_fit.predict(len(data),len(data)+6)
print(yhat)
```
